# Customizing Seaborn plots

1. darkgrid
2. white grid
3. dark
4. white
5. ticks

```python
# set_style() is to set style of figure
# set_style('dark')
```

```python
sns.lineplot(x='sepal.length',y='sepal.width',data=df)
plt.xlim(5)
sns.set_style('white')
plt.show()
```

```python
sns.lineplot(x='sepal.length',y='sepal.width',data=df)
plt.xlim(5)
sns.set_style('darkgrid')
plt.show()
```

```python
sns.lineplot(x='sepal.length',y='sepal.width',data=df)
plt.xlim(5)
sns.set_style('dark')
plt.show()
```

```python
sns.lineplot(x='sepal.length',y='sepal.width',data=df)
plt.xlim(5)
sns.set_style('ticks')
plt.show()
```

# subplots

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(4,2))
x = np.linspace(0,10,1000)
y = np.sin(x)
z = np.cos(x)
plt.subplot(1,2,1)
sns.lineplot(x=x,y=y,color='r',linewidth=4)
plt.subplot(1,2,2)
sns.lineplot(x=x,y=z,color='g',linewidth=6)
sns.set_style('darkgrid')
plt.show()
```

# FacetGrid

```python
# this helps in visualizing distribution of one variable as well as the relation
#relationship between multiple variables separately within subsets of your dataset
#using multiple panels
```

1. it takes dataframe as input and the names of the variables that will 
   form the row,column dimensions of the grid.
2. the variables should be categorical and the data at each level of the 
   variable will be used for facet along that axis

```python
plot = sns.FacetGrid(df,col='variety')
plot.map(plt.plot,'sepal.width')
plt.show()
```

```python
plot = sns.FacetGrid(df,col='variety')
plot.map(plt.plot,'petal.width')
plt.show()
```

# Distplot

```python
# it plots distribution of data along with histogram
plt.figure(figsize=(4,2))
sns.distplot(df['sepal.width'],color='r')
plt.show()
```

```python
plt.figure(figsize=(4,2))
sns.distplot(df['petal.width'],color='g')
plt.show()
```
