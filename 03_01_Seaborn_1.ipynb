# Data Visualization with Seaborn

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.describe()
```

```python
df.fillna(value={'age':df['age'].median()},inplace=True) 
```

```python
df.drop('Unnamed: 15',axis=1,inplace= True)
```

```python
m1=df['deck'].mode()
m2=df['embarked'].mode()
m3 = df['embark_town'].mode()
```

```python
df.fillna(value={'deck':m1[0],'embarked':m2[0],'embark_town':m3[0]},inplace=True)
```

# Line Charts

```python
plt.style.use('seaborn-darkgrid')
sns.lineplot(x=df1['engine_size'],y = df1['peak_rpm'].astype(float),data = df1)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn-darkgrid')
x = np.linspace(1,10,1000)
y = np.sin(x)
z = np.cos(x)
sns.lineplot(x=x,y=y)
sns.lineplot(x,z)
plt.show()
```

```python
col1 = np.linspace(0,10,1000)
col2 = np.sin(col1)
col3 = np.cos(col1)
df2 = pd.DataFrame({'C1':col1,'C2':col2,'C3':col3})
```

```python
sns.lineplot(x=df2.C1,y=df2.C2,data=df2)
sns.lineplot(x=df2.C1,y=df2.C3,data=df2)
plt.show()
```

```python
sns.lineplot(x=df2.C1,y=df2.C2,data=df2,color='r')
sns.lineplot(x=df2.C1,y=df2.C3,data=df2,color='y')
plt.show()
```

```python
sns.lineplot(x=df2.C1,y=df2.C2,data=df2,color='r',linewidth=4)
sns.lineplot(x=df2.C1,y=df2.C3,data=df2,color='y',linewidth=6)
plt.show()
```
