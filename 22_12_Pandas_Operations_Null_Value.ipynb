# Dealing with null values

```python
dframe = pd.DataFrame(M,index =dates)
dframe.columns = ['C1','C2','C3','C4','C5','C6','C7']
dframe
```

```python
# set null values in C5,C6,C7 colums in 0:8,0:2,5:6 row positions
```

```python
dframe.at[0:8,'C5'] = np.NaN
dframe.at[0:2,'C6'] = np.NaN
dframe.at[5:6,'C7'] = np.NaN
```

```python
dframe
```

```python
dframe.notna()
```

```python
dframe.isna()
```

```python
# fill all null values with 500
```

```python
dframe = dframe.fillna(500)
```

```python
dframe
```

```python
dframe.at[0:8,'C5'] = np.NaN
dframe.at[0:2,'C6'] = np.NaN
dframe.at[5:6,'C7'] = np.NaN
# replace null values in C6 with 150
```
